```python
# imdb_DF = pd.read_csv('Raw_Data/IMDb movies.csv')
imdb_DF = pd.read_csv('Raw_Data/clean_imdb.csv')
tomato_DF = pd.read_csv('Raw_Data/rotten_tomatoes_movies.csv')
```

```python
imdb_DF.columns
```

```python
#remove the columns we don't want for the merged ratings table 
mod_imdb = imdb_DF.drop(columns=['Unnamed: 0','metascore', 'reviews_from_users','reviews_from_critics'])
#renaming some columns
mod_imdb = mod_imdb.rename(columns = {'description':'imdb_description','avg_vote':'imdb rating'})
mod_imdb['year'] = mod_imdb['year'].astype(str)
```

```python
tomato_DF.columns
```

```python
tomato_DF=tomato_DF.drop(labels=0, axis=0)
tomato_DF.head()
```

```python
#remove columns 
mod_tomato = tomato_DF.drop(columns =['movie_info','genres', 'directors', 'authors','actors', 'runtime','production_company'])
mod_tomato = mod_tomato.rename(columns = {'movie_title':'title'})
```

```python
mod_tomato['year']=pd.DatetimeIndex(mod_tomato['original_release_date']).year
mod_tomato['year'] = mod_tomato['year'].astype(str).replace('\.0','',regex=True)
mod_tomato.head()
```

```python
temp_imdb = mod_imdb.drop(columns=['title'])
#renaming some columns
temp_imdb = temp_imdb.rename(columns = {'original_title':'title'})
```

```python
m1 = pd.merge(mod_tomato, temp_imdb, how='inner', on=['title','year'])
m2 = pd.merge(mod_tomato, mod_imdb, how='inner', on=['title','year'])
m3 = pd.concat([m1,m2]).drop_duplicates()
```
