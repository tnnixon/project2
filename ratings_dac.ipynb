```python
import pandas as pd
```

```python
IMDB_DF = pd.read_csv("IMDb_movies.csv")
IMDB_DF.head()
        
```

```python
RT_DF = pd.read_csv("rotten_tomatoes_movies.csv")
RT_DF.head()
```

```python
DF=IMDB_DF.merge(RT_DF, how='inner', left_on='original_title', right_on='movie_title',suffixes=('_imdb','_rt'))
DF.head()
```

```python
Ratings = DF[['title','year','original_release_date','avg_vote','tomatometer_rating','audience_rating']]
Ratings.head()
```

```python
Ratings = Ratings.rename(columns= {'avg_vote':'imdb_rating'})
Ratings.head()
```

```python
Ratings['rt_year'] = pd.DatetimeIndex(Ratings['original_release_date']).year
Ratings['rt_year'] = Ratings['rt_year'].fillna(0)
Ratings['rt_year'] = Ratings['rt_year'].astype(int)
del Ratings['original_release_date']
Ratings.head()
```

```python
StarBorn = Ratings.loc[Ratings["title"] == "A Star Is Born"]
StarBorn.head()
```

```python
StarBorn = Ratings.loc[(Ratings["title"] == "A Star Is Born")&(Ratings['year']== Ratings['rt_year'])]
StarBorn.head()
```

```python
neworder = ['title','year','rt_year','imdb_rating','tomatometer_rating','audience_rating']
Ratings=Ratings.reindex(columns=neworder)
Ratings.head()
```

```python
# Create SQL table using SQL Alchemy? No file created.
from sqlalchemy import create_engine
engine = create_engine('sqlite:///ratings2.sqlite', echo=False)
Ratings.to_sql('ratings2', con=engine, if_exists='append')
```
